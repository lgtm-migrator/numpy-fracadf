# Load Packages

```python
import sys
sys.path.append('..')

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
from numpy_fracadf import fracadf
from numpy_fracdiff import fracdiff
```

# Load Data

```python
with np.load('data/demo1.npz') as data:
    X = data['px']

X = X[1:]  # chop 01-Jan
```

# Find best fractal order

```python
alpha = 0.01
n_trunc = 200

%time d = fracadf(X, alpha=alpha, n_trunc=n_trunc)

print(d)
```

# Display results

```python
for j in range(X.shape[1]):
    z = fracdiff(X[:, j], order=d[j], truncation=n_trunc)
    plt.plot(z[n_trunc:]);
    plt.show();
```
